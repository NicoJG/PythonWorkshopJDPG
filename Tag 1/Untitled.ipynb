```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import animation

fig,ax = plt.subplots()
lines, = plt.plot([],[], 'o')

ax.axis("scaled")
ax.set_xlim(-1,1)
ax.set_ylim(-1,1)

fps = 60

x = 0
y = 0

v_x = np.random.rand()*2-1
v_y = np.random.rand()*2-1

def animieren(frame):
    global x,y,v_x,v_y
    #t = frame/fps
    x += v_x/fps
    y += v_y/fps
    if (x < -1) or (x > 1):
        v_x = -v_x
    if (y < -1) or (y > 1):
        v_y = -v_y
    
    lines.set_data(x,y)
    return lines,

#Hier darfst du natÃ¼rlich frames und interval so einstellen, wie du mÃ¶chtest
anim = animation.FuncAnimation(fig, animieren, interval = 1000/fps, blit=True)

plt.show()
```

```python
import matplotlib.pyplot as plt
from numpy import pi

n = 16
# create a n x n square with a marker at each point
x_data = []
y_data = []
for x in range(n):
    for y in range(n):
        x_data.append(x)
        y_data.append(y)

fig,ax = plt.subplots(figsize=[7,7])

# important part:
# calculate the marker size so that the markers touch

# radius in data coordinates:
r = 0.5 
# radius in display coordinates:
r_ = ax.transData.transform([r,0])[0] - ax.transData.transform([0,0])[0]
# marker size as the area of a circle
marker_size = pi * r_**2

ax.scatter(x_data, y_data, s=marker_size, edgecolors='black')

plt.show()
```

```python
import matplotlib.pyplot as plt
from numpy import pi

n = 4
# create a n x n square with a marker at each point as dummy data
x_data = []
y_data = []
for x in range(n):
    for y in range(n):
        x_data.append(x)
        y_data.append(y)

# open figure
fig,ax = plt.subplots(figsize=[10,10])
# set limits BEFORE plotting
ax.set_xlim((0,n-1))
ax.set_ylim((0,n-1))
# radius in data coordinates:
r = 0.5 # units
# radius in display coordinates:
r_ = ax.transData.transform([r,0])[0] - ax.transData.transform([0,0])[0] # points
# marker size as the area of a circle
marker_size = (2*r_)**2
# plot
ax.scatter(x_data, y_data, s=marker_size)

plt.show()
```
