```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import animation

fig,ax = plt.subplots()
lines, = plt.plot([],[], 'o')

ax.axis("scaled")
ax.set_xlim(-1,1)
ax.set_ylim(-1,1)

fps = 60

x = 0
y = 0

v_x = np.random.rand()*2-1
v_y = np.random.rand()*2-1

def animieren(frame):
    global x,y,v_x,v_y
    #t = frame/fps
    x += v_x/fps
    y += v_y/fps
    if (x < -1) or (x > 1):
        v_x = -v_x
    if (y < -1) or (y > 1):
        v_y = -v_y
    
    lines.set_data(x,y)
    return lines,

#Hier darfst du natÃ¼rlich frames und interval so einstellen, wie du mÃ¶chtest
anim = animation.FuncAnimation(fig, animieren, interval = 1000/fps, blit=True)

plt.show()
```
