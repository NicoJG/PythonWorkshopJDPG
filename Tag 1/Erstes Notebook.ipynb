```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
from matplotlib.animation import FuncAnimation
```

```python
%matplotlib notebook
```

```python
fig = plt.figure()
lines = [plt.plot([],'-')[0] for i in range(2)]

plt.axis("scaled")
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

x = np.linspace(0,2*np.pi,100)
def animieren(frame):
    x = np.sin(2*np.pi*frame/100)
    y = np.cos(2*np.pi*frame/100)
    x2 = 0.5*np.cos(np.pi+2*np.pi*frame/100)
    y2 = 0.5*np.sin(np.pi+2*np.pi*frame/100)
    
    lines[0].set_data((0,x),(0,y))
    lines[1].set_data((0,x2),(0,y2))
    return line
    
anim = FuncAnimation(fig, animieren, interval=20, blit=True)
plt.show()
```
